### 3. Inference (10 Points)

Implement a deep learning inference detection model into a deployment [Single Image; Video; and Live Segmentation].

```python
## Image Testing

import pixellib
from pixellib.instance import custom_segmentation

segment_image = custom_segmentation()
segment_image.inferConfig(num_classes=2, class_names= ["BG", "papercup", "plasticcup"])
segment_image.load_model("/content/drive/Shareddrives/CpE403_Final/mask_rcnn_models/mask_rcnn_model.030-0.057511.h5")
segment_image.segmentImage("/content/drive/Shareddrives/CpE403_Final/Testing & Inference/srcimg_paper.jpg", show_bboxes = True, output_image_name = "/content/drive/Shareddrives/CpE403_Final/Testing & Inference/outimg_paper.jpg")
segment_image.segmentImage("/content/drive/Shareddrives/CpE403_Final/Testing & Inference/srcimg_plastic.jpg", show_bboxes = True, output_image_name = "/content/drive/Shareddrives/CpE403_Final/Testing & Inference/outimg_plastic.jpg")
```

```python
## Video Testing

import pixellib
from pixellib.instance import custom_segmentation

segment_video = custom_segmentation()
segment_video.inferConfig(num_classes= 2, class_names= ["BG", "papercup", "plasticcup"])
segment_video.load_model("/content/drive/Shareddrives/CpE403_Final/mask_rcnn_models/mask_rcnn_model.030-0.057511.h5")
segment_video.process_video("/content/drive/Shareddrives/CpE403_Final/Testing & Inference/srcvid_paper.mp4", show_bboxes=True, frames_per_second = 15, output_video_name="/content/drive/Shareddrives/CpE403_Final/Testing & Inference/outvid_paper.mp4")
segment_video.process_video("/content/drive/Shareddrives/CpE403_Final/Testing & Inference/srcvid_plastic.mp4", show_bboxes=True, frames_per_second = 15, output_video_name="/content/drive/Shareddrives/CpE403_Final/Testing & Inference/outvid_plastic.mp4")
```
